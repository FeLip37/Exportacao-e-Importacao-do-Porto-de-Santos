```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

df = pd.read_csv('EXP_IMP.csv', sep=';')

df.columns = df.columns.str.strip().str.replace('\r', '')

df['Fluxo'] = df['Fluxo'].str.strip().str.replace('\r', '')

df_pivot = df.pivot(index='Ano', columns='Fluxo', values='Valor US$ FOB')

font = {
    'family': 'serif',
    'color': 'black',
    'weight': 'normal',
    'size': 14,
}

ax = df_pivot.plot(kind='bar', color={'Exportação': 'blue', 'Importação': 'orange'}, alpha=0.8)

ax.spines[['top', 'right']].set_visible(False)

plt.title('Exportação e Importação do Porto de Santos', fontdict=font)
plt.ylabel('Valor US$ FOB', fontdict=font)
plt.xlabel('Ano', fontdict=font)

formatter = FuncFormatter(lambda x, _: f'{x/1e9:,.0f}B')
ax.yaxis.set_major_formatter(formatter)

plt.legend(title='Tipo')
plt.xticks(rotation=0)
plt.tight_layout()

plt.savefig('export_import_plot.png')
```
